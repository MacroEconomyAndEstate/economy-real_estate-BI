```python
import pandas as pd

df2023 = pd.read_csv('행정구역_읍면동_별_5세별_주민등록인구_2011년__20231206205006.csv')

date = df2023.pop('시점').reset_index().drop('index', axis=1)
date = date.iloc[2:, :].copy()
date = date.reset_index().drop('index', axis=1)
# print(date.shape) --> (4, 1)


regions = []
for i in range(0, df2023.shape[1], 22):
    region_data = df2023.iloc[:, i:i+22].copy()
    regions.append(region_data)

                              
# regions 리스트에 저장된 데이터를 각각의 region 변수에 할당
for i, region_data in enumerate(regions, start=0):
    globals()['region{}'.format(i)] = region_data

lst = ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', 
       '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원특별자치도', 
        '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']

for i in range(len(lst)):
    globals()['region{}'.format(i)].columns = globals()['region{}'.format(i)].iloc[1]
    globals()['region{}'.format(i)] = globals()['region{}'.format(i)][2:].reset_index(drop=True)
    globals()['region{}'.format(i)]['행정구역'] = lst[i]
    globals()['region{}'.format(i)] = pd.concat([globals()['region{}'.format(i)]['행정구역'], 
                                                 globals()['region{}'.format(i)].drop('행정구역', axis=1)], axis=1)
for i in range(len(lst)):
    globals()['region{}'.format(i)] = pd.concat([date, globals()['region{}'.format(i)]], axis=1)
df = [globals()['region{}'.format(i)] for i in range(len(lst))]
result2023 = pd.concat(df, ignore_index=True)
```
