```python
quarter_to_month = {'Q1': '03-01', 'Q2': '06-01', 'Q3': '09-01', 'Q4': '12-01'}
def convert_to_datetime(df, cycle):
    if cycle == 'Q':
        return pd.to_datetime(df.str[:4] + '-' + df.str[4:].map(quarter_to_month), format='%Y-%m-%d')
    elif cycle == 'M':
        return pd.to_datetime(df.str[:4] + '-' + df.str[4:] + '-01', format='%Y-%m-%d')
    elif cycle == 'D':
        return pd.to_datetime(df.str[:4] + '-' + df.str[4:6] + '-' + df.str[6:], format='%Y-%m-%d')
    elif cycle == 'A':
        return pd.to_datetime(df.str[:4] + '-12-01', format='%Y-%m-%d')
    
def return_final_df(df, cycle, var_name):
    df['TIME'] = convert_to_datetime(df['TIME'], cycle)
    df.rename(columns={"DATA_VALUE":var_name}, inplace=True)
    df = pd.DataFrame(df, columns=["TIME", var_name])
    df.set_index("TIME", drop=True, append=False, inplace=True)
    return df
```
